2. Using Bayes' theorem, calculate the probability that an individual actually has the marker given a positive test result.

#### Answer 
1. Let the events be defined as follows: <br>
$ M $: The individual has the genetic marker. <br>
$ -M $: The individual does not have the genetic marker. <br>
$ T^+ $: The test result is positive. <br> <br>
The given probabilities are: <br>
$ P(M) = 0.3 $: Probability that an individual has the marker. <br>
$ P(-M) = 1 - P(M) = 0.7 $: Probability that an individual does not have the marker. <br>
$ P(T^+ \mid M) = 0.95 $: Probability of a positive test result given the individual has the marker (true positive rate). <br>
$ P(T^+ \mid -M) = 1 - 0.8 = 0.2 $: Probability of a positive test result given the individual does not have the marker (false positive rate). <br> <br>
2. Bayes' theorem states: <br>
$ P(M \mid T^+) = \frac{P(T^+ \mid M) P(M)}{P(T^+)} $ <br>
To calculate $ P(T^+) $, the total probability of a positive test result, we use the law of total probability: <br>
$ P(T^+) = P(T^+ \mid M) P(M) + P(T^+ \mid -M) P(-M) $ <br>
$ P(T^+) = (0.95)(0.3) + (0.2)(0.7) $ <br>
$ P(T^+) = 0.285 + 0.14 = 0.425 $ <br>
Now, use Bayes' theorem: <br>
$ P(M \mid T^+) = \frac{(0.95)(0.3)}{0.425} $ <br>
$ P(M \mid T^+) = \frac{0.285}{0.425} \approx 0.6706 $ <br>
Thus, the probability that an individual actually has the marker given a positive test result is approximately: <br>
$ P(M \mid T^+) \approx 67.06\% $ <br>

$ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}, $ <br>

Substitute $ n = 12 $, $ p = \frac{1}{6} $, and $ k = 4 $: <br>
