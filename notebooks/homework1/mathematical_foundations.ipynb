### Question 1
Let $ f(x) = x^3 - 6x^2 + 9x - 4 $. 
1. Find $ f'(x) $, the first derivative of $ f(x) $.
2. Determine the critical points of $ f(x) $ by solving $ f'(x) = 0 $.
3. Using the second derivative test, classify each critical point as a local maximum, minimum, or point of inflection.


### Answer 1
Let $ f(x) = x^3 - 6x^2 + 9x - 4 $. 
1. $ f'(x) = 3x^2 - 12x + 9 $.  
2. Solving $ f'(x) = 0 $:  
   $ 3x^2 - 12x + 9 = 0 $  
   $ x^2 - 4x + 3 = 0 $  
   $ (x - 3)(x - 1) = 0 $, so $ x = 1 $ and $ x = 3 $ are the critical points.  
3. The second derivative is $ f''(x) = 6x - 12 $.  
   - At $ x = 1 $, $ f''(1) = 6(1) - 12 = -6 $ (local maximum).  
   - At $ x = 3 $, $ f''(3) = 6(3) - 12 = 6 $ (local minimum).

### Question 4
Let $ g(x) = x e^x $.  
1. Find $ g'(x) $ using the product rule.  
2. Determine the integral $ \int x e^x \, dx $ by using integration by parts.

### Answer 4 
1. $ g'(x) = e^x + xe^x $ (Product Rule).  
2. Using integration by parts:  
   Let $ u = x $ and $ dv = e^x \, dx $.  
   Then $ du = dx $ and $ v = e^x $.  
   $$
   \int x e^x \, dx = x e^x - \int e^x \, dx = x e^x - e^x + C = e^x(x - 1) + C
   $$

### Question 6
Consider the following system of linear equations:  
$$
2x + 3y - z = 4
$$
$$
x - 2y + 4z = -3  
$$
$$
3x + y + 2z = 5
$$
1. Write the system in matrix form $ AX = B $.  
2. Find the determinant of $ A $.  
3. If the determinant is non-zero, use the inverse of $ A $ to solve for $ X $.


### Answer 6
1. Matrix form:  
   $$
   A = \begin{pmatrix} 2 & 3 & -1 \\ 1 & -2 & 4 \\ 3 & 1 & 2 \end{pmatrix}, \quad 
   X = \begin{pmatrix} x \\ y \\ z \end{pmatrix}, \quad 
   B = \begin{pmatrix} 4 \\ -3 \\ 5 \end{pmatrix}
   $$
2. Determinant of $ A $:  
   $$
   \text{det}(A) = 2(-2 \cdot 2 - 4 \cdot 1) - 3(1 \cdot 2 - 4 \cdot 3) + (-1)(1 \cdot 1 - (-2) \cdot 3) = 7
   $$
3. Since $ \text{det}(A) \neq 0 $, $ A $ is invertible, and $ X = A^{-1}B $.

### Question 9

A matrix $ D $ is defined as:

$$
D = \begin{pmatrix} a & 0 & 0 \\ 0 & b & 0 \\ 0 & 0 & c \end{pmatrix}
$$

1. Describe the type of matrix $ D $ and explain its properties.  
2. If $ D $ is a diagonal matrix, find its determinant.  
3. Explain the significance of a diagonal matrix in terms of linear transformations.


### Answer 9

A matrix $ D $ is defined as:

$$
D = \begin{pmatrix} a & 0 & 0 \\ 0 & b & 0 \\ 0 & 0 & c \end{pmatrix}
$$

1. $ D $ is a diagonal matrix. A diagonal matrix has non-zero entries only on the main diagonal, and all off-diagonal elements are zero. Its properties include:
   - Easy to compute the determinant and inverse.
   - Eigenvalues are the diagonal entries.

2. $$
\text{det}(D) = a \cdot b \cdot c
$$

3. Diagonal matrices represent scaling transformations along the coordinate axes. Each diagonal entry corresponds to the scaling factor in the direction of the respective axis.


### Question 10

Let $ E $ be a $ 2 \times 2 $ matrix defined as:

$$
E = \begin{pmatrix} 3 & 4 \\ 2 & 1 \end{pmatrix}
$$

1. Find the eigenvalues of $ E $.  
2. For each eigenvalue, find the corresponding eigenvector.  
3. Explain how the eigenvalues and eigenvectors can be used to understand transformations represented by $ E $.

### Answer 10

Let $ E $ be a $ 2 \times 2 $ matrix defined as:

$$
E = \begin{pmatrix} 3 & 4 \\ 2 & 1 \end{pmatrix}
$$

1. The eigenvalues are found by solving $ \text{det}(E - \lambda I) = 0 $:

$$
\text{det} \begin{pmatrix} 3 - \lambda & 4 \\ 2 & 1 - \lambda \end{pmatrix} = (3 - \lambda)(1 - \lambda) - 8 = \lambda^2 - 4\lambda - 5 = 0
$$

$$
\lambda = 5, -1
$$

2. - For $ \lambda = 5 $:

$$
(E - 5I) \mathbf{v} = 0 \implies \begin{pmatrix} -2 & 4 \\ 2 & -4 \end{pmatrix} \mathbf{v} = 0 \implies \mathbf{v} = \begin{pmatrix} 2 \\ 1 \end{pmatrix}
$$

- For $ \lambda = -1 $:

$$
(E + I) \mathbf{v} = 0 \implies \begin{pmatrix} 4 & 4 \\ 2 & 2 \end{pmatrix} \mathbf{v} = 0 \implies \mathbf{v} = \begin{pmatrix} -1 \\ 1 \end{pmatrix}
$$

3. Eigenvalues indicate the scaling factors in the direction of the eigenvectors. The eigenvectors represent the directions that remain invariant under the transformation $ E $.

### Question 11

In a population, 30% of individuals have a specific genetic marker. A test for this marker has a 95% accuracy rate for individuals with the marker and an 80% accuracy rate for those without the marker.

1. Define the events and write out the given probabilities.  
2. Using Bayes' theorem, calculate the probability that an individual actually has the marker given a positive test result.


### Answer 11

In a population, 30% of individuals have a specific genetic marker. A test for this marker has a 95% accuracy rate for individuals with the marker and an 80% accuracy rate for those without the marker.

1. Let:
- $ M $: Individual has the marker $( P(M) = 0.3 )  $
- $ M^c $: Individual does not have the marker $( P(M^c) = 0.7 )  $
- $ T^+ $: Positive test result  

Given:
- $ P(T^+|M) = 0.95 $
- $ P(T^+|M^c) = 0.2 $  
- $ P(T^-|M^c) = 0.2 $
  
2. $$
P(M|T^+) = \frac{P(T^+|M)P(M)}{P(T^+)} 
$$

Where:
$$
P(T^+) = P(T^+|M)P(M) + P(T^+|M^c)P(M^c)
$$

$$
P(T^+) = (0.95)(0.3) + (0.2)(0.7) = 0.285 + 0.14 = 0.425
$$

$$
P(M|T^+) = \frac{(0.95)(0.3)}{0.425} = \frac{0.285}{0.425} \approx 0.67
$$


### Question 12

A die is rolled 12 times. Let $ X $ be the random variable representing the number of times a "5" appears.

1. What type of probability distribution does $ X $ follow?  
2. Calculate the probability that "5" appears exactly 4 times.  
3. Determine the expected number of times "5" will appear in 12 rolls.

### Answer 12

A die is rolled 12 times. Let $ X $ be the random variable representing the number of times a "5" appears.

1. $ X $ follows a Binomial distribution: $ X \sim \text{Binomial}(n = 12, p = \frac{1}{6}) $.

2. $$
P(X = 4) = \binom{12}{4} \left( \frac{1}{6} \right)^4 \left( \frac{5}{6} \right)^8
$$

$$
P(X = 4) = 495 \cdot \left( \frac{1}{1296} \right) \cdot \left( 0.233 \right) \approx 0.089
$$

3. $$
E[X] = n \cdot p = 12 \cdot \frac{1}{6} = 2
$$


### Question 13

A bookstore averages 3 customer visits per hour. Assume the number of customer visits per hour follows a Poisson distribution.

1. Calculate the probability that exactly 5 customers visit in a given hour.  
2. Find the probability that at least 1 customer visits in a given hour.


### Answer 13

A bookstore averages 3 customer visits per hour. Assume the number of customer visits per hour follows a Poisson distribution.

1. $$
P(X = 5) = \frac{\lambda^5 e^{-\lambda}}{5!} = \frac{3^5 e^{-3}}{120} \approx 0.1008
$$

2. $$
P(X \geq 1) = 1 - P(X = 0) = 1 - \frac{3^0 e^{-3}}{0!} = 1 - e^{-3} \approx 0.9502
$$


### Question 14

### Answer 14

$$
P(45 \leq Y \leq 55) = P\left(-1 \leq Z \leq 1\right) \approx 0.6827
$$

y = \mu + z\sigma = 50 + (1.04)(5) = 55.2

### Question 15

A company's product quality control process involves two independent tests. Each test has a 98% probability of detecting a defect if it is present.

1. What is the probability that both tests detect a defect when it is present?  
2. What is the probability that at least one test detects a defect when it is present?  
3. If the company requires both tests to detect a defect before rejecting a product, calculate the probability that a defective product will be rejected.


### Answer 15

A company's product quality control process involves two independent tests. Each test has a 98% probability of detecting a defect if it is present.

1. $$
P(\text{Both detect}) = 0.98 \cdot 0.98 = 0.9604
$$

2. $$
P(\text{At least one detects}) = 1 - P(\text{Neither detects}) = 1 - (1 - 0.98)^2 = 1 - 0.0004 = 0.9996
$$

3. $$
P(\text{Rejected}) = P(\text{Both detect}) = 0.9604
$$
